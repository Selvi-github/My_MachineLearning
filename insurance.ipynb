```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
ins=pd.read_csv('/content/insurance_customer_data.csv')
```

```python
 a=ins.iloc[:,[2,3]].values
 print(a)
```

```python
wcss=[]
for i in range(1,11):
 kmeans=KMeans(n_clusters=i,init='k-means++',random_state=2)
 kmeans.fit(a)
 wcss.append(kmeans.inertia_)
```

```python
 kmeans=KMeans(n_clusters=3,init='k-means++',random_state=0)
 insurance=kmeans.fit_predict(a)
 print(insurance)
```

```python
 b=kmeans.fit_predict(a)
 ins['cluster']=b
```

```python
 plt.figure(figsize=(5,5))
 plt.scatter(a[b==0,0],a[b==0,1],s=50,c='pink',label='High')
 plt.scatter(a[b==2,0],a[b==2,1],s=50,c='yellow',label='Low')
 plt.scatter(a[b==1,0],a[b==1,1],s=50,c='gray',label='Medium')
 plt.scatter(a[b==3,0],a[b==3,1],s=50,c='brown',label='Extra high')
 plt.scatter(a[b==4,0],a[b==4,1],s=50,c='purple',label='Extra low')
 plt.title("Customer Insurance")
 plt.xlabel('Annual income')
 plt.ylabel('Spending score')
 plt.legend()
 plt.show()
```

```python
 v=pd.Series(insu).value_counts()
 plt.figure(figsize=(5,5))
 plt.pie(v,labels=v.index,autopct='%1.2f%%')
plt.title('Customer Insurance')
 plt.show()
```

```python
 sns.set()
 plt.plot(range(1,11),wcss)
 plt.title('The Elbow Method')
 plt.xlabel('Number of clusters')
 plt.ylabel('WCSS')
 plt.show()
```
